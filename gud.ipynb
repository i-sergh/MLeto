```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

from matplotlib import pyplot as plt
```

```python
(X_train, y_train), (X_valid, y_valid) = mnist.load_data()
```

```python
plt.figure(figsize=(5, 5))

for i in range(100):
    plt.subplot(10, 10, i+1)
    plt.imshow(X_train[i], cmap='Greys')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
X_train_L = X_train.reshape(60000, 784).astype('float32')
X_valid_L = X_valid.reshape(10000, 784).astype('float32')
```

```python
X_train_L /= 255
X_valid_L /= 255
```

```python
y_train = keras.utils.to_categorical(y_train, 10)
y_valid = keras.utils.to_categorical(y_valid, 10)
```

```python
model = Sequential()
model.add(Dense(64, activation='sigmoid', input_shape=(784,)))
model.add(Dense(10, activation='softmax'))
```

```python

model.compile(loss='mean_squared_error', optimizer=SGD(lr=0.01), metrics=['accuracy'])
```

```python
model.fit(X_train_L, y_train, batch_size=128, epochs=100, verbose=1, validation_data=(X_valid_L, y_valid))
```

```python
model2 = Sequential()
model2.add(Dense(800, activation='sigmoid', input_shape=(784,)))
model2.add(Dense(10, activation='softmax'))

model2.compile(loss='mean_squared_error', optimizer=SGD(lr=0.01), metrics=['accuracy'])
```

```python
model2.fit(X_train_L, y_train, batch_size=128, epochs=200, verbose=1, validation_data=(X_valid_L, y_valid))
```

```python
with open('model_params.json', 'w') as f:
    f.write(model.to_json())
model.save_weights('weights.h5')
```

```python
with open('model_params.json', 'r') as f:
    loaded_model = keras.models.model_from_json(f.read())
    
loaded_model.load_weights('weights.h5')
```

```python
load_test = loaded_model.predict(X_valid_L)
```

```python
for ind, par in enumerate(load_test[10]):
    print(par, y_valid[10][ind], ind, sep="==>")
 
plt.imshow(X_valid[ 10], cmap='plasma')
plt.show()
```
